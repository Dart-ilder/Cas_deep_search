```python
import torch
print("Pytorch " + torch.__version__)

import pandas as pd
import numpy as np

from torch import nn
from torch.utils.data import Dataset, DataLoader
import os
```

```python
with open('./cas_dataset_kira.tsv') as file:
    cas_voc = pd.read_csv(file, delimiter="\t", comment='=')
cas_voc.head(3)
```

```python
embs_list = []
for i in range(2465):
    path  = f"./batched_embs/batch{i}.txt"
    with open(path, "r") as f:
        batch = np.loadtxt(f)
    
    embs_list.append(batch)
    os.system('cls')
    print(f"{i}/2465")
    
embs = np.concatenate(embs_list, axis=0)
print(embs.shape)
```

```python
# Disassemble batches into one big 2d array
embs_list = [row for row in embs]
len(embs_list)
```

```python
embs_path = "embs_concatenated.txt"
np.savetxt(embs_path, np.array(embs_list), delimiter=',', newline='\n', header='', footer='', comments='# ')
```

```python
with open(embs_path, "r") as f:
    embs_str = f.readlines()
cas_voc["Embs"] = list(map(lambda x: x.strip("\n"), embs_str))
file = "./cas_voc.tsv"
cas_voc.to_csv(file, sep='\t', index=False)
```
