```python
import torch
print("Pytorch " + torch.__version__)

import pandas as pd
import numpy as np

from torch import nn
from torch.utils.data import Dataset, DataLoader
import os
```

```python
with open('./prot_voc_1000.tsv') as file:
    prot_voc = pd.read_csv(file, delimiter="\t", comment='=')
prot_voc.head(3)
```

```python
embs_list = []
for i in range(125):
    path  = f"./batched_embs/batch{i}.txt"
    with open(path, "r") as f:
        batch = np.loadtxt(f)
    
    embs_list.append(batch)
    print(f"{i}/125")
    
embs = np.concatenate(embs_list, axis=0)
print(embs.shape)
```

```python
# Disassemble batches into one big 2d array
embs_list = [row for row in embs]
len(embs_list)
```

```python
prot_voc["Embs"] = embs_list
```

```python
prot_voc["Embs"][0]
```

```python
embs_path = "embs_concatenated.txt"
np.savetxt(embs_path, np.array(embs_list), delimiter=',', newline='\n', header='', footer='', comments='# ')
```
