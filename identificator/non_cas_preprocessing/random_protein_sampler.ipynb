```python
import Bio.SeqIO as BSIO
import pandas as pd
archea = BSIO.parse("archea.xml", "uniprot-xml")
bacteria = BSIO.parse("bacteria.xml", "uniprot-xml")
prot_voc = pd.DataFrame()
length = 355599 #sum(1 for _ in archea) + sum(1 for _ in bacteria)
print(length)
k = 0
for i in archea:
    row = pd.DataFrame({"Gene_id": i.id, "Gene_family": "Junk", "Prot": str(i.seq)}, index=[0])
    prot_voc = pd.concat([prot_voc, row], axis=0, ignore_index=True)
    k += 1
    print("{:.2%}".format(k/length))

for i in bacteria:
    row = pd.DataFrame({"Gene_id": i.id, "Gene_family": "Junk", "Prot": str(i.seq)}, index=[0])
    prot_voc = pd.concat([prot_voc, row], axis=0, ignore_index=True)
    k += 1
    print("{:.2%}".format(k/length))
prot_voc.head(3)
```
